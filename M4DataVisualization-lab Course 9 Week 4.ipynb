## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0, 0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = "SELECT ConvertedComp FROM master"

df = pd.read_sql_query(QUERY, conn)

df['ConvertedComp'].plot(kind='hist', edgecolor='green')

print(df)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = "SELECT Age FROM master"

df = pd.read_sql_query(QUERY, conn)

df['Age'].plot(kind='box', color='red', vert=False)

print(df)
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = "SELECT Age, WorkWeekHrs FROM master"

df = pd.read_sql_query(QUERY, conn)

df.boxplot(column=['Age', 'WorkWeekHrs'], vert=False, patch_artist=True, color= 'blue')

print()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"

df = pd.read_sql_query(QUERY, conn)
df.plot.scatter(x='WorkWeekHrs', y='CodeRevHrs', s=df['Age'], alpha=0.5, edgecolor= "yellow")

print(df)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = "SELECT Respondent, SurveyEase FROM master LIMIT 5"

# Read data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plot a pie chart using pandas
df['Respondent'].value_counts().plot.pie(autopct='%1.0f%%', legend=False, title='Survey Respondent Distribution')

# Print the DataFrame
print(df)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)

median_data = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

median_data.plot(kind='bar', stacked=True, colormap='viridis')
print(median_data)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)

median_comp_by_age = df.groupby('Age')['ConvertedComp'].median()
median_comp_by_age.plot(kind='line', marker='*', color='orange', linestyle='-', linewidth=2)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(*) AS Count
FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY, conn)

ax = df.plot.barh(x='MainBranch', y='Count', color='blue')
ax.set_title('Distribution of MainBranch')
ax.set_xlabel('Count')
ax.set_ylabel('MainBranch')
ax.grid(axis='x')

ax.figure.show()
```
